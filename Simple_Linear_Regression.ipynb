```python
# Simple linear regression calculatings 
from scipy import stats
import numpy as np
from matplotlib import pyplot as plt

x = np.array([112, 335, 168, 278, 321, 526, 462, 383, 584])
y = np.array([1251, 2351, 1962, 2135, 2121, 4522, 3634, 2966, 5412])

slope, intercept, r_value, p_value, std_err = stats.linregress(x,y)

plt.plot(x, y, 'ro', color = 'red')
plt.plot(x, x*slope + intercept)
plt.ylabel('Price')
plt.xlabel('Size')
plt.axis([0, 600, 0, 6000])
plt.show()
```
